```python
# Clean up lowercase, strip spaces
df[8] = df[8].str.lower().str.strip()

# Standardize similar resolution names
df[8] = df[8].replace({
    'full hd': '1080p',
    'fhd+': '1080p',
    'hd+': '720p',
    'hd ready': '720p',
    'uhd': '4k',
    'ultra hd': '4k',
    'retina': 'retina',
    '2k': '1440p',
    '2.2k': '1440p',
    '2,2k': '1440p',
    '2.5k': '1440p',
    '3k': '3k',
    '3,2k': '3k',
    '4k': '4k',
    '5k': '5k'
})

# Rename the column for clarity
df = df.rename(columns={8: 'Screen_Resolution'})

# preview the result
df['Screen_Resolution'].value_counts(dropna=False).head(20)
```

```python
# Group resolutions into general categories
def categorize_resolution(res):
    if pd.isna(res):
        return 'unknown'
    elif res in ['720p']:
        return 'HD'
    elif res in ['1080p']:
        return 'Full HD'
    elif res in ['1440p', 'qhd', 'qhd+', '2.2k', '2.5k', '2.8k']:
        return 'QHD'
    elif res in ['3k', 'retina']:
        return '3K'
    elif res in ['4k', 'uhd', 'uhd+']:
        return '4K'
    elif res in ['5k']:
        return '5K'
    else:
        return 'Other'

df['Resolution_Category'] = df['Screen_Resolution'].apply(categorize_resolution)

# check result
df['Resolution_Category'].value_counts(dropna=False)
```
